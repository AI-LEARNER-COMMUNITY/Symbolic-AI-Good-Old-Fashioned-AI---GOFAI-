Good Old-Fashioned AI (GOFAI), also known as Symbolic AI, refers to a traditional approach to artificial intelligence that dominated the field in the early decades of its development. This approach is based on symbolic reasoning and manipulation of symbols to represent knowledge and solve problems.

Symbolic AI is an approach to artificial intelligence that emerged in the mid-20th century. It focuses on representing knowledge using symbols and employing formal logic for reasoning and decision-making. Unlike modern machine learning approaches that rely on statistical patterns, Symbolic AI emphasizes explicit programming of rules and knowledge structures.

# Knowledge Representation:
In this approach, knowledge is explicitly represented using symbols, typically in the form of rules, logic, and symbols. These symbols are manipulated through various algorithms to perform tasks.Here, We'll provide a simple example using a text-based approach with natural language processing (NLP) techniques.

```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Define some example knowledge
knowledge_base = [
    {"text": "Cats are mammals.", "label": "animal"},
    {"text": "Dolphins are intelligent marine animals.", "label": "animal"},
    {"text": "Python is a programming language.", "label": "technology"},
    {"text": "Machine learning involves training models on data.", "label": "technology"},
]

# Separate texts and labels
texts = [item["text"] for item in knowledge_base]
labels = [item["label"] for item in knowledge_base]

# Create a pipeline with a CountVectorizer and a Multinomial Naive Bayes classifier
model = make_pipeline(CountVectorizer(), MultinomialNB())

# Train the model on the knowledge base
model.fit(texts, labels)

# Test the model with new data
new_data = ["Cheetahs are the fastest land animals.", "Java is a programming language."]
predicted_labels = model.predict(new_data)

# Print the predictions
for text, label in zip(new_data, predicted_labels):
    print(f"Text: {text} | Predicted Label: {label}")
```

This above machine learning code using a bag-of-words approach and a Naive Bayes classifier. It represents knowledge about animals and technology and predicts the category of new statements.

# Logic and Rule-Based Systems:
Symbolic AI heavily relies on formal logic and rule-based systems. Logical reasoning and deduction are used to draw conclusions from the knowledge encoded in the system. In this pythone notebook, to create a rule-based system for classifying animals based on their characteristics. We'll define a few rules and implement a simple inference engine.

In this example, we create a RuleBasedSystem class that allows us to define rules and infer conclusions based on input facts. The rules are simple conditions and conclusions related to animal characteristics.

```python
class RuleBasedSystem:
    def __init__(self):
        self.rules = []

    def add_rule(self, condition, conclusion):
        self.rules.append({"condition": condition, "conclusion": conclusion})

    def infer(self, facts):
        for rule in self.rules:
            condition_met = all(facts.get(feature) == value for feature, value in rule["condition"].items())
            if condition_met:
                return rule["conclusion"]
        return None

# Example Rule-Based System for Animal Classification
animal_classifier = RuleBasedSystem()

# Define rules
animal_classifier.add_rule({"has_fur": True, "has_claws": True}, "Mammal")
animal_classifier.add_rule({"lives_in_water": True, "has_fins": True}, "Fish")
animal_classifier.add_rule({"can_fly": True}, "Bird")

# Test the rule-based system
facts1 = {"has_fur": True, "has_claws": True}
facts2 = {"lives_in_water": True, "has_fins": True}
facts3 = {"can_fly": True}

result1 = animal_classifier.infer(facts1)
result2 = animal_classifier.infer(facts2)
result3 = animal_classifier.infer(facts3)

# Print the results
print(f"Facts: {facts1} | Inferred Conclusion: {result1}")
print(f"Facts: {facts2} | Inferred Conclusion: {result2}")
print(f"Facts: {facts3} | Inferred Conclusion: {result3}")
```

# Expert Systems:

Expert systems are a notable application of Symbolic AI. These systems attempt to emulate the decision-making abilities of a human expert in a particular domain by encoding their knowledge and decision-making processes in a set of rules.

Creating a complete expert system involves defining rules, encoding domain knowledge, and implementing an inference engine. Below is a Machine learning code example using Python to create an expert system for diagnosing medical conditions based on symptoms. We'll define rules for a few medical conditions and implement an inference engine to diagnose a patient based on their symptoms.

```python
class ExpertSystem:
    def __init__(self):
        self.rules = []

    def add_rule(self, condition, diagnosis):
        self.rules.append({"condition": condition, "diagnosis": diagnosis})

    def diagnose(self, symptoms):
        possible_diagnoses = []

        for rule in self.rules:
            condition_met = all(symptoms.get(symptom) == value for symptom, value in rule["condition"].items())
            if condition_met:
                possible_diagnoses.append(rule["diagnosis"])

        return possible_diagnoses

# Example Expert System for Medical Diagnosis
medical_expert = ExpertSystem()

# Define rules for medical conditions
medical_expert.add_rule({"fever": True, "cough": True}, "Flu")
medical_expert.add_rule({"headache": True, "nausea": True}, "Migraine")
medical_expert.add_rule({"rash": True, "itching": True}, "Allergic Reaction")

# Test the expert system with patient symptoms
patient_symptoms = {"fever": True, "cough": True, "headache": False, "nausea": False, "rash": False, "itching": False}

diagnoses = medical_expert.diagnose(patient_symptoms)

# Print the diagnosis results
if diagnoses:
    print(f"Possible Diagnoses for the Patient: {', '.join(diagnoses)}")
else:
    print("No specific diagnosis found based on the symptoms.")
```

In this above example ML code, we create an ExpertSystem class with methods to add rules and diagnose patients based on symptoms. We define rules for three medical conditions (Flu, Migraine, and Allergic Reaction) and test the expert system with a set of patient symptoms.

# Inference Engines:
In Symbolic AI, inference engines are used to draw logical conclusions from the available knowledge base. These engines follow predefined rules and logical principles to infer new information.

Creating a complete inference engine involves defining rules, encoding knowledge, and implementing the reasoning process. Below is a Machine learning code using Python to create a basic inference engine that evaluates logical rules.

```python
class InferenceEngine:
    def __init__(self):
        self.knowledge_base = []

    def add_rule(self, antecedent, consequent):
        self.knowledge_base.append({"antecedent": antecedent, "consequent": consequent})

    def infer(self, facts):
        inferred_conclusions = []

        for rule in self.knowledge_base:
            antecedent_met = all(facts.get(feature) == value for feature, value in rule["antecedent"].items())
            if antecedent_met:
                inferred_conclusions.append(rule["consequent"])

        return inferred_conclusions

# Example Inference Engine
inference_engine = InferenceEngine()

# Add rules to the knowledge base
inference_engine.add_rule({"has_wings": True, "can_fly": True}, "Bird")
inference_engine.add_rule({"lives_in_water": True, "has_fins": True}, "Fish")
inference_engine.add_rule({"has_fur": True, "climbs_trees": True}, "Mammal")

# Test the inference engine with facts
input_facts = {"has_wings": True, "can_fly": True, "lives_in_water": False, "has_fins": False, "has_fur": False, "climbs_trees": False}

inferred_results = inference_engine.infer(input_facts)

# Print the inferred conclusions
if inferred_results:
    print(f"Inferred Conclusions: {', '.join(inferred_results)}")
else:
    print("No conclusions inferred based on the given facts.")
```

# Limited Adaptability:
GOFAI systems are often criticized for their lack of adaptability and inability to learn from experience. These systems rely on explicitly programmed rules and do not have the capacity for learning or adapting to new information in the way modern machine learning systems do.

To illustrate the limited adaptability of a GOFAI (Good Old-Fashioned AI) system can involve a scenario where the system encounters new information but is unable to adapt its rules. Below is an example using a basic rule-based system for animal classification, where the system cannot adapt to new animals it hasn't encountered before.

```python
class RuleBasedAnimalClassifier:
    def __init__(self):
        self.rules = []

    def add_rule(self, features, label):
        self.rules.append({"features": features, "label": label})

    def classify(self, features):
        for rule in self.rules:
            if all(features.get(feature) == value for feature, value in rule["features"].items()):
                return rule["label"]
        return "Unknown"

# Create a rule-based animal classifier
animal_classifier = RuleBasedAnimalClassifier()

# Add rules for known animals
animal_classifier.add_rule({"has_fur": True, "has_claws": True}, "Mammal")
animal_classifier.add_rule({"lives_in_water": True, "has_fins": True}, "Fish")
animal_classifier.add_rule({"can_fly": True}, "Bird")

# Test the classifier with known and unknown animals
known_animal_features = {"has_fur": True, "has_claws": True}
unknown_animal_features = {"has_feathers": True, "can_swim": True}

known_animal_label = animal_classifier.classify(known_animal_features)
unknown_animal_label = animal_classifier.classify(unknown_animal_features)

# Print the classification results
print(f"Known Animal Classification: {known_animal_label}")
print(f"Unknown Animal Classification: {unknown_animal_label}")
```

In this above example, the RuleBasedAnimalClassifier has rules for known animals (Mammal, Fish, and Bird). When given the features of a known animal, it can classify it correctly. However, when presented with features of an unknown animal, it labels it as "Unknown" because the system lacks the ability to adapt to new rules for animals it hasn't encountered before.
